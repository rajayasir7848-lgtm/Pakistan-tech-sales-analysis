```python
city_revenue =df.groupby('City')['Total_Revenue'].sum().sort_values(ascending=False)
city_revenue.plot(kind='bar',color='skyblue',edgecolor='black')
plt.title('PER CITY REVENUE')
plt.xlabel('CITY')
plt.ylabel('Total_revenue')
plt.tight_layout()
plt.savefig('PER CITY REVENUE.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
# Revenue by Product
```

```python
revenue_product= df.groupby('Product')['Total_Revenue'].sum().sort_values(ascending=False)
revenue_product.plot(kind='bar',color='blue',edgecolor='black')
plt.title('REVENUE BY PRODUCT')
plt.xlabel('PRODUCT')
plt.ylabel('TOTAL REVENUE')
plt.tight_layout()
plt.xticks(rotation=45,ha='right')
plt.savefig('PRODUCT REVENUE.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
revenue_product= df.groupby('Salesperson')['Total_Revenue'].sum().sort_values(ascending=False)
revenue_product.plot(kind='bar',color='blue',edgecolor='black')
plt.title('REVENUE BY Salesperson')
plt.xlabel('Salesperson')
plt.ylabel('TOTAL REVENUE')
plt.tight_layout()

plt.xticks(rotation=45,ha='right')
plt.savefig('REVENUE BY SALES PERSON.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
daily_revenue = df.groupby('Date')['Total_Revenue'].sum()
daily_revenue.plot(kind='line', marker='o', color='green')
plt.title('Daily Revenue Trend')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.savefig('dailytrend.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
figure,axes=plt.subplots(2,2,figsize=(12,8))
city_revenue =df.groupby('City')['Total_Revenue'].sum()
axes[0][0].bar(city_revenue.index,city_revenue.values,color='skyblue')
axes[0][0].set_title('Revenue by City')
#product_revenue
product_revenue=df.groupby('Product')['Total_Revenue'].sum().sort_values(ascending=False)
axes[0][1].bar(product_revenue.index,product_revenue.values,color='blue')
axes[0][1].set_title('REVENUE BY PRODUCT')
#Salesperson Revenue
Salesperson_Revenue= df.groupby('Salesperson')['Total_Revenue'].sum().sort_values(ascending=False)
axes[1][0].bar(Salesperson_Revenue.index,Salesperson_Revenue.values)
axes[1][0].set_title('Salesperson_Revenue')
daily_trend = df.groupby('Date')['Total_Revenue'].sum()
axes[1][1].plot(daily_trend.index,daily_trend .values,marker='o',color='red')
axes[1][1].set_title('Daily Revenue Trend')

plt.tight_layout()
plt.savefig('dashboard.png',dpi=300,bbox_inches='tight')
plt.show;
```

```python
import pandas as pd

with pd.ExcelWriter('pakistan_sales_report.xlsx', engine='openpyxl') as writer:
    # Sheet 1 - Cleaned Data
    df.to_excel(writer, sheet_name='Cleaned Data', index=False)
    
    # Sheet 2 - Revenue by City
    df.groupby('City')['Total_Revenue'].sum().sort_values(ascending=False).to_excel(writer, sheet_name='Revenue by City')
    
    # Sheet 3 - Revenue by Product
    df.groupby('Product')['Total_Revenue'].sum().sort_values(ascending=False).to_excel(writer, sheet_name='Revenue by Product')
    
    # Sheet 4 - Top Salesperson
    df.groupby('Salesperson')['Total_Revenue'].sum().sort_values(ascending=False).to_excel(writer, sheet_name='Top Salesperson')

print("Excel recreated successfully!")
```
